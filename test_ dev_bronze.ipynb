```python
""" Test and Development notebook to get the API requests and storage it into a json file"""
from pyspark.sql import SparkSession
import requests
import json
import os
```

```python
""" Create Spark Session """
spark = SparkSession.builder.appName("BronzeLayer").getOrCreate()

API_URL = "https://api.openbrewerydb.org/breweries"

""" Create the data path """
bronze_path = os.getcwd()
os.makedirs(bronze_path, exist_ok=True)
```

```python
"""Test the API """
def test_api_request(url):
  """Function to check if there is the correct answer from API request"""
  try:
    response = requests.get(url)
    assert response.status_code == 200, f"Error: {response.status_code}"
    data = response.json()
    assert isinstance(data, list), "Error: Response is not a correct type"
    print("Test has been concluded with success")
  except Exception as e:
    print(f"Error {e}")
```

```python
"""Run the Test function"""
test_api_request(API_URL)
```

```python
""" Extract data from API """
response = requests.get(API_URL)
data = response.json()
print(data)
```

```python
""" Save data into json file"""
with open(f"{bronze_path}/dev_bronze_layer.json", "w") as f:
    json.dump(data, f, indent=4)
print("Dados extraÃ­dos e armazenados na camada Bronze.")
```
